```python
 # ilgilenecegimiz sutuunlar ve sayisallastirma islemleri
relevant_columns = ["Overall", "Computer", "Preparation", "Income",
                    "Gender", "Hometown", "Gaming", "Attendance", "Extra"]
# 1. Income
income_numeric_list = []
for income_str in df['Income']:
    income_str_cleaned = str(income_str).strip()
    if income_str_cleaned == "Low (Below 15,000)":
        income_numeric_list.append(0)
    elif income_str_cleaned == "Lower middle (15,000-30,000)":
        income_numeric_list.append(1)
    elif income_str_cleaned == "Upper middle (30,000-50,000)":
        income_numeric_list.append(2)
    elif income_str_cleaned == "High (Above 50,000)":
        income_numeric_list.append(3)
    else:
        income_numeric_list.append(np.nan) #null icin
df['Income_n'] = income_numeric_list

# 2. Preparation (Ders Çalışma Süresi)
preparation_numeric_list = []
for prep_str in df['Preparation']:
    prep_str_cleaned = str(prep_str).strip()
    if prep_str_cleaned == "0-1 Hour":
        preparation_numeric_list.append(0)
    elif prep_str_cleaned == "2-3 Hours":
        preparation_numeric_list.append(1)
    elif prep_str_cleaned == "More than 3 Hours":
        preparation_numeric_list.append(2)
    else:
        preparation_numeric_list.append(np.nan)
df['Preparation_n'] = preparation_numeric_list

# 3. Gaming (Oyun Oynama Süresi)
gaming_numeric_list = []
for game_str in df['Gaming']:
    game_str_cleaned = str(game_str).strip()
    if game_str_cleaned == "0-1 Hour":
        gaming_numeric_list.append(0)
    elif game_str_cleaned == "2-3 Hours":
        gaming_numeric_list.append(1)
    elif game_str_cleaned == "More than 3 Hours":
        gaming_numeric_list.append(2)
    else:
        gaming_numeric_list.append(np.nan)
df['Gaming_n'] = gaming_numeric_list

# 4. Attendance (Devam Durumu)
# "Below 40%" -> 0
# "40%-59%" -> 1
# "60%-79%" -> 2
# "80%-100%" -> 3
attendance_numeric_list = []
for att_str in df['Attendance']:
    att_str_cleaned = str(att_str).strip()
    if att_str_cleaned == "Below 40%":
        attendance_numeric_list.append(0)
    elif att_str_cleaned == "40%-59%":
        attendance_numeric_list.append(1)
    elif att_str_cleaned == "60%-79%":
        attendance_numeric_list.append(2)
    elif att_str_cleaned == "80%-100%":
        attendance_numeric_list.append(3)
    else:
        attendance_numeric_list.append(np.nan)
df['Attendance_n'] = attendance_numeric_list

# 5. Gender (Cinsiyet)
gender_numeric_list = []
for gender_str in df['Gender']:
    gender_str_cleaned = str(gender_str).strip()
    if gender_str_cleaned == "Female":
        gender_numeric_list.append(0)
    elif gender_str_cleaned == "Male":
        gender_numeric_list.append(1)
    else:
        gender_numeric_list.append(np.nan)
df['Gender_Numeric'] = gender_numeric_list

# 6. Hometown (Memleket)
hometown_numeric_list = []
for home_str in df['Hometown']:
    home_str_cleaned = str(home_str).strip()
    if home_str_cleaned == "Village":
        hometown_numeric_list.append(0)
    elif home_str_cleaned == "City":
        hometown_numeric_list.append(1)
    else:
        hometown_numeric_list.append(np.nan)
df['Hometown_n'] = hometown_numeric_list

# Extra (Ekstra Etkinlik)

extra_numeric_list = []
for extra_str in df['Extra']:
    extra_str_cleaned = str(extra_str).strip()
    if extra_str_cleaned == "No":
        extra_numeric_list.append(0)
    elif extra_str_cleaned == "Yes":
        extra_numeric_list.append(1)
    else:
        extra_numeric_list.append(np.nan)
df['Extra_n'] = extra_numeric_list

# 8. Computer (Bilgisayar Yeterliliği) Sütunu
# zaten sayisal deger var. kontrol edelim
# Veri tipini kontrol edip, eğer object ise int'e çevirelim.
if df['Computer'].dtype == 'object':
    df['Computer_n'] = pd.to_numeric(df['Computer'], errors='coerce') # Hatalı girişleri NaN yapar
else:
    df['Computer_n'] = df['Computer'] # Zaten sayısal ise direkt ata
```
